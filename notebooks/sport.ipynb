```python
# Étape 1A — Renommer la clé Ticket_id → Ticket_id_full

ages_ren = ages.rename(columns={"Ticket_id": "Ticket_id_full"})
genders_ren = genders.rename(columns={"Ticket_id": "Ticket_id_full"})
origin_ren = origin.rename(columns={"Ticket_id": "Ticket_id_full"})

# Étape 1B — Fusionner les dimensions démographiques sur Ticket_id_full

dim_demographics = ages_ren.merge(genders_ren, on="Ticket_id_full", how="inner") \
                           .merge(origin_ren, on="Ticket_id_full", how="inner")

# Étape 1C — Vérification

print(dim_demographics.head())
print(dim_demographics.info())
print(dim_demographics.nunique())
```

```python
# On repart de ta table scans existante

FACT_SCANS = scans.copy()

# Étape 1 : Créer la colonne Date
FACT_SCANS["Date"] = FACT_SCANS["Timestamp"].dt.date

# Étape 2 : Dériver les composantes temporelles
FACT_SCANS["Hour"] = FACT_SCANS["Timestamp"].dt.hour
FACT_SCANS["Minute"] = FACT_SCANS["Timestamp"].dt.minute

# Étape 3 : Création de Time_bin_10min (créneau de 10 minutes)
FACT_SCANS["Time_bin_10min"] = (FACT_SCANS["Hour"] * 60 + FACT_SCANS["Minute"]) // 10 * 10

# Étape 4 : Ajouter la mesure (1 scan = 1 ligne = 1 événement)
FACT_SCANS["Scan_count"] = 1

# Étape 5 : Ne garder que les colonnes nécessaires
FACT_SCANS = FACT_SCANS[[
    "Ticket_id_full",
    "Venue",
    "Date",
    "Timestamp",
    "Hour",
    "Minute",
    "Time_bin_10min",
    "Scan_count"
]]

# Affichage
print(FACT_SCANS.head())
print(FACT_SCANS.info())
```

```python
DIM_VENUE = venue_mapping.copy()

# ------------------------------
# 1) Renommer les colonnes
# ------------------------------
DIM_VENUE = DIM_VENUE.rename(columns={
    "Venue Capacity (nb of seats)": "Venue_capacity",
    "Day of competition": "Day_of_competition",
    "Doors opening": "Doors_opening",
    "Start of the session": "Start_of_session",
    "End of the session": "End_of_session",
    "Doors closing": "Doors_closing"
})

# ------------------------------
# 2) Séparer Localization en Latitude / Longitude
# ------------------------------

# Exemple de format : "48.84713743431346, 2.250150141915884"
DIM_VENUE["Latitude"] = DIM_VENUE["Localization"].apply(lambda x: float(str(x).split(",")[0].strip()))
DIM_VENUE["Longitude"] = DIM_VENUE["Localization"].apply(lambda x: float(str(x).split(",")[1].strip()))

# On peut ensuite supprimer la colonne d'origine si tu veux :
DIM_VENUE = DIM_VENUE.drop(columns=["Localization"])

# ------------------------------
# 3) Calculer la durée de session
# ------------------------------
# On convertit les heures en datetime (au même jour)
import pandas as pd

# Convertir les strings en datetime (heure seulement)
DIM_VENUE["Start_of_session_dt"] = pd.to_datetime(DIM_VENUE["Start_of_session"], format="%I.%M %p")
DIM_VENUE["End_of_session_dt"] = pd.to_datetime(DIM_VENUE["End_of_session"], format="%I.%M %p")

# Calcul de la durée
DIM_VENUE["Session_duration_minutes"] = (
    (DIM_VENUE["End_of_session_dt"] - DIM_VENUE["Start_of_session_dt"])
    .dt.total_seconds() / 60
)

# On peut supprimer les colonnes intermédiaires
DIM_VENUE = DIM_VENUE.drop(columns=["Start_of_session_dt", "End_of_session_dt"])

# ------------------------------
# 4) Afficher le résultat
# ------------------------------
print(DIM_VENUE.head())
print(DIM_VENUE.info())
```

```python
import pandas as pd

# Étape 1 — récupérer toutes les dates présentes dans FACT_SCANS
date_min = pd.to_datetime(FACT_SCANS["Date"]).min()
date_max = pd.to_datetime(FACT_SCANS["Date"]).max()

# Étape 2 — générer un calendrier complet de date_min à date_max
DIM_DATE = pd.DataFrame({"Date": pd.date_range(date_min, date_max, freq="D")})

# Étape 3 — ajouter les attributs temporels
DIM_DATE["Year"] = DIM_DATE["Date"].dt.year
DIM_DATE["Month"] = DIM_DATE["Date"].dt.month
DIM_DATE["Month_name"] = DIM_DATE["Date"].dt.month_name()
DIM_DATE["Day"] = DIM_DATE["Date"].dt.day
DIM_DATE["Weekday"] = DIM_DATE["Date"].dt.weekday + 1     # 1 = Lundi, 7 = Dimanche
DIM_DATE["Weekday_name"] = DIM_DATE["Date"].dt.day_name()
DIM_DATE["IsWeekend"] = DIM_DATE["Weekday"].isin([6, 7]).astype(int)

# Affichage
print(DIM_DATE.head())
print(DIM_DATE.info())
```
